### Solve Problem

- Từ các đồ thì trên ta có thể lên ý tưởng là tìm được số cực tiểu/ cực đại của đồ thị từng cột, và lấy số cực tiểu/cực đại phổ biến nhất


```python
from scipy.ndimage import median_filter
import pandas as pd
import numpy as np
```

```python
def find_rep(my_list, window_size):

    # Làm mềm danh sách
    my_list = median_filter(my_list, size=window_size)
    
    mean_value = np.mean(my_list)
    check = 1
    if my_list[0] > mean_value:
        check = 1
    else:
        check = 0
    cnt = 0
    for i in range(1, len(my_list)):
        if my_list[i] > mean_value and check == 0:
            cnt += 1
            check = 1
        elif my_list[i] < mean_value and check == 1:
            cnt += 1
            check = 0
    return cnt//2
```

```python
def rep_counting(path_file):
    df = pd.read_csv(path_file)
    maybe_rep = []
    for i in range(1, 9):
        for j in range(1, 10):
            maybe_rep.append(find_rep(df[str(i)], j))
    
#     print(maybe_rep)
    rep = max(maybe_rep, key=maybe_rep.count)
    
    return rep
```

### Result

```python
print("Kết quả là:", rep_counting('A-5678-12.csv')) # 13 mới đúng
```
